```python
db = client['sample_supplies']
```

```python
sales = db['sales']
```

```python
l = list(sales.find({}))
```

```python
customer_age = []
for i in l:
    customer_age.append(i['customer']['age'])
customer_age
```

```python
count = {"15-20":0,"20-25":0,"25-30":0,"30-35":0,"35-40":0,"40-45":0,"45-50":0,"50-55":0,"55-60":0,"60-65":0,"65-70":0,"70-75":0,"75-80":0}
```

```python
for i in customer_age:
    if 15<=i<20:
        count['15-20'] += 1
    elif i<25:
        count['20-25'] += 1
    elif i<30:
        count['25-30'] += 1
    elif i<35:
        count['30-35'] += 1
    elif i<40:
        count['35-40'] += 1
    elif i<45:
        count['40-45'] += 1
    elif i<50:
        count['45-50'] += 1
    elif i<55:
        count['50-55'] += 1
    elif i<60:
        count['55-60'] += 1
    elif i<65:
        count['60-65'] += 1
    elif i<70:
        count['65-70'] += 1
    elif i<75:
        count['70-75'] += 1
    else:
        count['75-80'] += 1
count
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
count_items = count.items()
final = pd.DataFrame(list(count_items),columns=['age',"num"])
final.head()
```

```python
plt.plot(list(final['age']),list(final['num']))
plt.title("Distribution of Customer Ages")
plt.xlabel("customer.age")
plt.xticks(rotation=90)
plt.show()
```
